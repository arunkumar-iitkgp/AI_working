```python
import torch
import torchvision
model = torchvision.models.resnet18(weights='DEFAULT')
import torchvision.transforms as transforms
preprocess = transforms.Compose(
    [
        transforms.Resize(256),
        transforms.CenterCrop(224),
        transforms.ToTensor(),
        transforms.Normalize(
            mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]
        ),
    ]
)

from PIL import Image

# Replace with the actual path to your image file
img = Image.open('IMG_0322.jpg')
img_preprocessed = preprocess(img)
model.eval()
with torch.no_grad():
    outputs = model(img_preprocessed.unsqueeze(0))
_, predicted = torch.max(outputs, 1)
print(f"Predicted Class Index: {predicted.item()}")
import json
import urllib.request

url = "https://raw.githubusercontent.com/anishathalye/imagenet-simple-labels/master/imagenet-simple-labels.json"
urllib.request.urlretrieve(url, "imagenet-simple-labels.json")

with open("imagenet-simple-labels.json") as f:
    class_idx = json.load(f)

predicted_index = 834
label = class_idx[predicted_index]
print(f"The label for index {predicted_index} is: {label}")
```
